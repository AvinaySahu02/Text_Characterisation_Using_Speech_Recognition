**Google Web Speech API:**

This is a free service provided by Google i.e. Recognize Google, through which we can convert audio speech files to text and perform operations on them

```python
with sr.AudioFile('/content/drive/MyDrive/Deeep Learning Project/DL Projects/Text Characterization using Speech Recognition/Code and Files/voice-data.wav') as source:
    audio_file = recognizer.record(source)
    result=recognizer.recognize_google(audio_data=audio_file)
```

There are two types of taind input:

1)Duration: This is used to select a specific time of audio data i.e. If you wnat to select just 5 seconds then you can set Duration=5 and it will only select the 5 seconds of audio file

```python
with sr.AudioFile('/content/drive/MyDrive/Deeep Learning Project/DL Projects/Text Characterization using Speech Recognition/Code and Files/voice-data.wav') as source:
    audio_file_duration = recognizer.record(source, duration=10.0)
    result1=recognizer.recognize_google(audio_data=audio_file_duration)
```

```python
result1
```

2)Offset: This is used when you cut out some part of the starting data from your audio file i.e. if you don't want first 2 seconds then you can select offset as 2 and it willl stop the first two seconds.

```python
with sr.AudioFile('/content/drive/MyDrive/Deeep Learning Project/DL Projects/Text Characterization using Speech Recognition/Code and Files/voice-data.wav') as source:
    audio_file_offset = recognizer.record(source, offset=5.0)
    result2=recognizer.recognize_google(audio_data=audio_file_offset)
```

```python
result2
```

Combining both the file the duration and the offset

```python
with sr.AudioFile('/content/drive/MyDrive/Deeep Learning Project/DL Projects/Text Characterization using Speech Recognition/Code and Files/voice-data.wav') as source:
    audio_file_combine = recognizer.record(source, duration=10.0, offset=5.0)
    result3=recognizer.recognize_google(audio_data=audio_file_combine)
```

**Count the repetition of the words**

First we will store the unique words in a dictionary

```python
#To count the number of times the unique words apper, first in the unique_word lust
word_dict = {} #An empth dictionary
for word in result_str:
    word_dict[word] = 0
print(word_dict)
```

```python
for word in result_str:
    word_dict[word] = word_dict[word] + 1
print('The count for each word spoken number of times are: ',word_dict)
```

```python
cols = ['Repetition']
count_df = pd.DataFrame.from_dict(word_dict, orient='index', columns=cols)
```
